```python
month = pd.get_dummies((df['Month']))
month.head(5)
```

```python
visitor_type = pd.get_dummies(df['VisitorType'])
visitor_type.head(5)
```

```python
data = df
data = pd.concat([data, month, visitor_type], axis = 1)
data.drop(['Month', 'VisitorType'], axis=1, inplace=True)
data.head(5)
```

```python
# Logistic Regression

X = data.drop(['Revenue'], axis=1)
y = data['Revenue']

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
prediction = log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, prediction, output_dict=True)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prediction)
```

# KNN

```python

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

# Feature scaling

```python


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
classification_report(y_test, y_pred, output_dict=True)
```

# Random Forest

```python


from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(X_train, y_train)
```

```python
prediction = rf.predict(X_test)
```

```python
confusion_matrix(y_test, prediction)
```

```python
accuracy_score(y_test, prediction)
```

```python
classification_report(y_test, prediction, output_dict=True)
```

# Oversampling

### Random forest on oversampled data

```python
X_train.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 2)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
```

```python
X_train_res.shape
```

```python
rf.fit(X_train_res, y_train_res)
prediction = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test, prediction))
```

```python
(classification_report(y_test, prediction, output_dict=True))
```

```python
print(accuracy_score(y_test, prediction))
```

### KNN on oversampled data

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train_res, y_train_res)
y_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
